Dari histogram dan informasi di atas, dataset ini dapat dikatakan balance karena dilihat dari target variabel yaitu Growth_Milestone, jumlah nilai 0 dan jumlah nilai 1 tidak berbeda jauh, hanya berbeda 1. Jadi, dataset ini masih tergolong balanced dataset.

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Soil_Type', hue='Growth_Milestone', data=df, palette={0: '#4fa3ff', 1: '#ff9900'})
plt.title('Soil Type vs Growth Milestone')
plt.xlabel('Soil Type')
plt.ylabel('Count')
plt.legend(title='Growth Milestone', loc='upper right')
plt.show()
```

Dari chart di atas, dapat dilihat bahwa tipe tanah loam yang yang tanamannya banyak mencapai target pertumbuhan. Selain itu, tanaman yang gagal mencapai target pertumbuhan terbanyak biasanya ditanam di tanah bertipe clay.

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Water_Frequency', hue='Growth_Milestone', data=df, palette={0: '#4fa3ff', 1: '#ff9900'})
plt.title('Water Frequency vs Growth Milestone')
plt.xlabel('Water Frequency')
plt.ylabel('Count')
plt.legend(title='Growth Milestone', loc='upper left')
plt.show()
```

Dari chart di atas, dapat dilihat bahwa frekuensi penyiraman tanaman yang membuat tanaman mencapai target pertumbuhan terbanyak adalah daily atau harian. Namun, frekuensi harian jugalah tanaman banyak yang tidak mencapai target pertumbuhan.

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Fertilizer_Type', hue='Growth_Milestone', data=df, palette={0: '#4fa3ff', 1: '#ff9900'})
plt.title('Fertilizer vs Growth Milestone')
plt.xlabel('Fertilizer')
plt.ylabel('Count')
plt.legend(title='Growth Milestone', loc='upper left')
plt.show()
```

Dari chart di atas, dapat dilihat bahwa tanaman yang banyak mencapai target pertumbuhan biasanya menggunakan pupuk tipe organik. Banyak tanaman yang gagal mencapai target pertumbuhan dengan tidak memakai pupuk sama sekali.

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Soil_Type', hue='Fertilizer_Type', data=df[df['Growth_Milestone'] == 0], palette={'none': '#4fa3ff', 'chemical': '#ff9900', 'organic': '#ff6361'})
plt.title('Soil Type dengan Fertilizer untuk Growth Milestone = 0')
plt.xlabel('Soil Type')
plt.ylabel('Count')
plt.legend(title='Fertilizer', loc='upper right')
plt.show()
```

Dari chart di atas, dapat dilihat bahwa tanaman banyak yang tidak mencapai target pertumbuhan apabila ditanam pada tanah bertipe sandy yang tidak menggunakan pupuk sama sekali. Selain itu dapat diartikan juga bahwa tanah bertipe sandy sangat buruk apabila tidak menggunakan pupuk. Perbedaannya cukup signifikan apabila pupuk digunakan untuk tanah sandy. Pupuk organik merupakan pupuk terbaik untuk tanah sandy.

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Soil_Type', hue='Fertilizer_Type', data=df[df['Growth_Milestone'] == 1], palette={'none': '#4fa3ff', 'chemical': '#ff9900', 'organic': '#ff6361'})
plt.title('Soil Type dengan Fertilizer untuk Growth Milestone = 1')
plt.xlabel('Soil Type')
plt.ylabel('Count')
plt.legend(title='Fertilizer', loc='upper left')
plt.show()
```

Dari chart di atas, dapat dilihat bahwa tanaman banyak mencapai target pertumbuhan apabila ditanam pada tanah bertipe loam yang pupuk organik. Selain itu, dapat diartikan juga bahwa pupuk organik paling bagus digunakan pada tanah dengan tipe loam.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sunlight_Hours', y='Temperature', hue='Growth_Milestone', data=df, palette={0: 'blue', 1: 'orange'})
plt.legend(title='Growth Milestone', loc='upper left')
plt.title('Sunlight Hours vs Temperature')
```

Dari scatter plot di atas, relasi antara sunlight hours dengan temperatur tidak terlalu memengaruhi target pertumbuhannya. Tidak ada pola yang jelas dan nilai 0 dan 1 tersebar di semua rentang, sehingga sunlight hours dan temperature tidak memiliki relasi yang kuat dan tidak terlalu memengaruhi pertumbuhan tanaman.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Humidity', y='Temperature', hue='Growth_Milestone', data=df, palette={0: 'blue', 1: 'orange'})
plt.legend(title='Growth Milestone', loc='upper left')
plt.title('Humidity vs Temperature')
```

Dari scatter plot ini, humidity dan temperature juga bukan merupakan relasi yang kuat yang dapat membuat perubahan yang kuat untuk pertumbuhan tanaman karena tidak ada pola yang jelas dan nilai 0 dan 1 nya tersebar di semua rentang.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sunlight_Hours', y='Humidity', hue='Growth_Milestone', data=df, palette={0: 'blue', 1: 'orange'})
plt.legend(title='Growth Milestone', loc='upper left')
plt.title('Sunlight Hours vs Humidity')
```

Dari scatter plot di atas, relasi antara sunlight hours dengan humidity juga tidak terlalu memengaruhi target pertumbuhannya. Sama seperti scatter plot sebelumnya, nilai growth milestone baik 0 ataupun 1 tersebar ke rentang manapun dan juga tidak ada pola yang jelas yang menandakan bahwa hubungan kedua fitur ini benar-benar memengaruhi pertumbuhan tanaman.
